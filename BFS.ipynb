```python
import collections


graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E'],
    'C': ['A', 'D', 'E', 'F'],
    'D': ['A', 'C', 'F'],
    'E': ['B', 'C'],
    'F': ['C', 'D']
}

def bfs(g, root):
    queue = collections.deque([root])
    visited = []

    while queue:
        node = queue.popleft()

        if node not in visited:
            visited.append(node)


            if node == 'F':
                break

            for item in g.get(node, []):
                if item not in visited:
                    queue.append(item)

    return visited


start_state = 'A'
order_of_nodes_visited = bfs(graph, start_state)

print(order_of_nodes_visited)
```
