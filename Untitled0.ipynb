```python
import pandas as pd

# Define base path
base_path = "/content/drive/My Drive/Database CSVs/"

```

```python
import csv

good_lines = []
file_path = base_path + "kiva_loans.csv"
output_path = base_path + "kiva_loans_clean.csv"

# Read and filter out malformed lines
with open(file_path, 'r', encoding='latin1', errors='replace') as infile:
    reader = csv.reader(infile)
    header = next(reader)
    expected_col_count = len(header)
    good_lines.append(header)

    for row in reader:
        if len(row) == expected_col_count:
            good_lines.append(row)

# Write cleaned version
with open(output_path, 'w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(good_lines)
```

```python
# Convert date column and extract time info
loans['date'] = pd.to_datetime(loans['date'], errors='coerce')
loans['year'] = loans['date'].dt.year
loans['month'] = loans['date'].dt.month
loans['quarter'] = loans['date'].dt.quarter
```

```python
# Create a simplified gender group column
def parse_gender(genders):
    if pd.isnull(genders):
        return 'Unknown'
    g_list = genders.lower().split(',')
    males = g_list.count('male')
    females = g_list.count('female')

    if males > 0 and females == 0:
        return 'Male Group' if males > 1 else 'Male'
    elif females > 0 and males == 0:
        return 'Female Group' if females > 1 else 'Female'
    elif males > 0 and females > 0:
        return 'Mixed Group'
    return 'Unknown'

loans['gender_group'] = loans['borrower_genders'].apply(parse_gender)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gender distribution by year
plt.figure(figsize=(12, 6))
sns.countplot(data=loans, x='year', hue='gender_group')
plt.title("Borrower Gender Distribution Over Years")
plt.xticks(rotation=45)
plt.show()

# Gender distribution by region
plt.figure(figsize=(12, 6))
top_regions = loans['region'].value_counts().nlargest(10).index
sns.countplot(data=loans[loans['region'].isin(top_regions)],
              y='region', hue='gender_group', order=top_regions)
plt.title("Top 10 Regions by Gender Group of Borrowers")
plt.show()
```

```python
# Estimated unmet need as expense proxy
loans['unfunded_amount'] = loans['loan_amount'] - loans['funded_amount']

# Average unfunded amount by year
plt.figure(figsize=(10, 5))
loans.groupby('year')['unfunded_amount'].mean().plot(kind='bar', color='orange')
plt.title("Average Unfunded Amount Per Year (Proxy for Expenses)")
plt.ylabel("Unfunded USD")
plt.xticks(rotation=45)
plt.show()
```

```python
region_expense = loans.groupby(['region', 'year'])['unfunded_amount'].mean().reset_index()
region_expense = region_expense[region_expense['unfunded_amount'] > 0]  # remove zero expenses
region_expense.sort_values(by='unfunded_amount', ascending=False).head(10)
```
